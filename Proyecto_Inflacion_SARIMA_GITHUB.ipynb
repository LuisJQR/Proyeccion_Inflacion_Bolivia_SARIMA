## Contexto Econ√≥mico Actual

En el contexto macroecon√≥mico que atraviesa actualmente Bolivia, caracterizado por **escasez de combustible**, un **tipo de cambio paralelo** significativamente por encima del oficial, **incremento de precios sobre bienes importados**, una **mayor emisi√≥n monetaria** y otros factores macroecon√≥micos, ha generado una *tendencia creciente en la inflaci√≥n, no observada en a√±os anteriores*.

Este fen√≥meno ha empezado a impactar en **la rentabilidad real de los portafolios de inversi√≥n**, calculada aproximadamente como:

$$
\text{Rentabilidad Real} \approx \text{Rentabilidad Nominal} - inflaci√≥n(\pi)
$$

Generando, en muchos casos, **spreads negativos** donde la rentabilidad del portafolio queda por debajo de la inflaci√≥n anual.

Ante esta situaci√≥n, los inversionistas han comenzado a priorizar instrumentos que ayuden a mitigar el impacto inflacionario. Entre ellos, destacan los **t√≠tulos indexados a la Unidad de Fomento a la Vivienda (UFV)**, los cuales ofrecen tasas nominales bajas (~0%) pero con un rendimiento **ajustado por inflaci√≥n** superior, lo que protege el poder adquisitivo del inversionista.

Como respuesta, el **Banco Central de Bolivia** ha incrementado la emisi√≥n de estos instrumentos, mientras que las **entidades financieras** han lanzado productos como **Dep√≥sitos a Plazo Fijo en UFV (DPF en UFV)** para captar liquidez, consolidando as√≠ una mayor oferta de instrumentos indexadas a las *UFV*.

### Uso del diccionarioo para convertirlos a fecha

Se debe trabajar con meses en ingles y usar ***pd.to_datetime***

```python
Dicc_Mes = {
    "Enero": "January",
    "Febrero": "February",
    "Marzo": "March",
    "Abril": "April",
    "Mayo": "May",
    "Junio": "June",
    "Julio": "July",
    "Agosto": "August",
    "Septiembre": "September",
    "Octubre": "October",
    "Noviembre": "November",
    "Diciembre": "December"
}
```

**Usar el formato correcto para los datos**

Nos aseguramos que al transponer los datos, sigan siendo del tipo decimal

```python
for col in Base_IPC.columns:
    if col != "MES":
        Base_IPC[col] = pd.to_numeric(Base_IPC[col], errors='coerce')
    
Base_IPC
```

**Transformaci√≥n de la base a formato columna**

Se realiza la transposici√≥n y limpieza de la base para obtener una columna unificada de IPC con su respectiva fecha.

```python
Base_IPC["MES"]=Base_IPC["MES"].replace(Dicc_Mes)

Base_IPC = Base_IPC.melt(id_vars="MES",var_name="A√ëO",value_name="IPC")

Base_IPC = Base_IPC[Base_IPC["IPC"]!=" "]

Base_IPC["Fecha"] = pd.to_datetime(Base_IPC["A√ëO"].astype(str) + "-" + Base_IPC["MES"],format="%Y-%B")

Base_IPC=Base_IPC.set_index("Fecha").sort_index()

Base_IPC.drop(columns=["MES","A√ëO"],inplace=True)
```

**CREAR DOS FUNCIONES**
1. Test_Est() &rarr; Nos ayudar√° a indicar si la Serie de Tiempo es **Estacionaria** o **No es Estacionaria**, a trav√©s, del Test Augmented Dickey-Fuller (Importante para aplicar SARIMA)
2. Graficar() &rarr; Graficar la **Serie Temporal**

```python
def Test_Est(Serie):
    Test_ADF = adfuller(Serie)
    if Test_ADF[1]<0.05:
        return "Estacionaria"
    else:
        return "No Estacionaria"
```

```python
def Graficar(Serie, Label,Color):
    plt.figure(figsize=(10,5))
    plt.plot(Serie.index, Serie, label=Label,color=Color)
    
    plt.xlabel("Fecha")
    plt.ylabel(Label)
    plt.title(Label)
    
    plt.show()
```

```python
Graficar(Base_IPC["IPC"],"IPC","darkblue")
```

### Consideraciones Importantes antes de An√°lizar las Series Temporales

La serie de tiempo debe ser estacionaria para poder aplicar SARIMA, esto quiere decir que la varianza y la media deben ser constantes. Por lo tanto, aplicaremos *Augmented Dickey-Fuller*.

En caso de no ser estacionaria, es debe a dos motivos:

1. La Serie Temporal presenta **Tendencia**, por lo tanto, se debe aplicar la diferencia **Œî** para eliminar la tendencia.
2. La Serie Temporal no presenta una **Varianza** constante, por lo tanto, se debe aplicar estrategias para estabilizar la varianza: Box-Cox, Log(), entre otros, para este proyecto usaremos Log() 

```python
print(f"La serie en Nivel es {Test_Est(Base_IPC['IPC'])}")
print(f"La serie en 1ra Diferencia es {Test_Est(Base_IPC['IPC'].diff().dropna())}")
print(f"La serie en 2da Diferencia es {Test_Est(Base_IPC['IPC'].diff().diff().dropna())}")
print(f"La serie en 3ra Diferencia es {Test_Est(Base_IPC['IPC'].diff().diff().diff().dropna())}")
#Sin embargo I(3) se pierde mucha precisi√≥n.
```

### An√°lisis

Se observa que el IPC no es estacionario en Nivel &rarr; I(0), por lo que es necesario aplicar diferencias. Sin embargo, se identifica un problema, la serie solo alcanza estacionariedad tras aplicar la tercera diferencia **ŒîŒîŒîIPC**, lo que implica una p√©rdida significativa de informaci√≥n. Debido a esto, antes de diferenciar en exceso, se optar√° por estabilizar la varianza para evaluar si esta transformaci√≥n mejora los resultados del test ADF.

```python
#Para estabilizar usaremos log()
Base_IPC["log_IPC"] = np.log(Base_IPC["IPC"])
Base_IPC["log_IPC"]
```

```python
print(f"La serie_log en Nivel es {Test_Est(Base_IPC['log_IPC'])}")
print(f"La serie_log en 1ra Diferencia es {Test_Est(Base_IPC['log_IPC'].diff().dropna())}")
#print(f"La serie_log en 2da Diferencia es {Test_Est(Base_IPC['log_IPC'].diff().diff().dropna())}")
```

### An√°lisis

Se observa que la serie es Estacionaria en primera diferencia, lo cual es recomendable para trabajar con SARIMA, al no perder informaci√≥n importante

```python
Base_IPC["D_log_IPC"] = Base_IPC['log_IPC'].diff()
Base_IPC.dropna(subset="D_log_IPC",inplace=True)
Base_IPC
```

### An√°lisis

Aplicaremos ***Autocorrelaci√≥n Parcial &rarr; AR(p)*** y ***Autocorrelaci√≥n &rarr; MA(q)*** para empezar a crear nuestro modelo ARIMA

```python
fig , axs = plt.subplots(1,2, figsize=(14,5))
plot_pacf(Base_IPC["D_log_IPC"],ax=axs[0])
plot_acf(Base_IPC["D_log_IPC"],ax=axs[1])
plt.show()
```

### An√°lisis

***PACF &rarr; 6 &rarr; AR(6)***

***ACF &rarr; 22 &rarr; MA(22)***


### Estacionalidad

Es conocido que la inflaci√≥n presenta un componente estacional en determinados meses, por ejemplo:

- ***Diciembre*** &rarr; Debido a las compras navide√±as la inflaci√≥n tiende a incrementar
- ***Enero*** &rarr; La inflaci√≥n tiende a disminuir

En ese sentido, es importante analizar la ***Estacionalidad*** de la ***Serie Temporal***, por lo tanto, vamos a descomponer la serie en:

$$
\text{IPC} = \text{Tendencia} + Estacionalidad + Residual
$$

```python
Descomposicion = seasonal_decompose (Base_IPC["IPC"],model ="additive")
Descomposicion.plot()
plt.tight_layout()
plt.show()
```

### An√°lisis

Se logra observar que la serie temporal tiene un componente estacional de s=12, por lo tanto, podriamos aplicar un

$$
\text{SARIMA (p,d,q) x (P,D,Q) s}
$$

### Estimaci√≥n del modelo SARIMA: enfoque autom√°tico vs manual

Vamos aplicar un AutoArima, este codigo nos permite realizar diferentes iteraciones para ***SARIMA (p,d,q) x (P,D,Q) s***, sin embargo, gracias a nuestro an√°lisis previo, sabemos que **s=12** y **d=1** y compararemos este modelo automatico, con el modelo manual de ARIMA (6,1,22) y ver cual es el mejor modelo en funci√≥n al ***Akaike (AIC)***

```python
Modelo_Manual = SARIMAX(Base_IPC["log_IPC"],
                        order=(6, 1, 22),
                        seasonal_order=(0, 0, 0, 0))

resultado = Modelo_Manual.fit(disp=False)

print(resultado.summary())
```

```python
Modelo_Auto = auto_arima(Base_IPC["log_IPC"],
    d=None, D=None, m=12,
    seasonal=True,
    stepwise=False,
    trace=True,
    error_action="ignore",
    suppress_warnings=True,
)

print(Modelo_Auto.summary())
```

### An√°lisis

Modelos encontrados:
- Modelo ***Auto_Arima &rarr; SARIMA(1,1,4)(0,0,0)[12]*** &rarr; ***AIC = -3470.685***
- Modelo Manual ***SARIMAX &rarr; ARIMA(6, 1, 22)*** &rarr; ***AIC = -3454.702*** (Adem√°s sus coeficientes no son significativos)

En ese sentido, el mejor modelo encontrado es por el ***Auto_Arima &rarr; SARIMA(1,1,4)(0,0,0)[12]***

### Proyecci√≥n Inflaci√≥n Anual

Procederemos a la proyecci√≥n de la inflaci√≥n anual para los siguientes 6 meses.

```python
Pronostico_Periodos = 6

Valor_Pronosticado_log, Intervalo_Confianza_log = Modelo_Auto.predict(n_periods=Pronostico_Periodos, return_conf_int=True)
```

### Transformar pron√≥stico de log(IPCs) a nivel original

```python
Valor_Pronosticado = np.exp(Valor_Pronosticado_log)
Intervalo_Confianza = np.exp(Intervalo_Confianza_log)
```

```python
Intervalo_Confianza
```

```python
Pronostico_Index = pd.date_range(start=Base_IPC.index[-1]+pd.DateOffset(months=1),periods=Pronostico_Periodos,freq="MS")
```

```python
plt.figure(figsize=(16,5))

#Base_filtrado = Base_IPC[Base_IPC.index > "2024-01-01"]

plt.plot(Base_IPC.index,Base_IPC["IPC"] , color="black" , label="IPC")
plt.plot(Pronostico_Index,Valor_Pronosticado , color="red" , label="Pronostico")
plt.fill_between(Pronostico_Index,Intervalo_Confianza[:,0],Intervalo_Confianza[:,1] , color="lightgreen",alpha=0.3 , label="Confianza al 95%")
plt.title("Proyecci√≥n del IPC Bolivia ‚Äì Pr√≥ximos 6 meses (SARIMA)")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
Base_Proyectado = pd.DataFrame({
    "IPC": Valor_Pronosticado,
}, index=Pronostico_Index)
Base_Proyectado
```

```python
Base_Total=Base_IPC.copy()
```

```python
Base_Total=pd.concat([Base_Total,Base_Proyectado])
```

```python
Base_Total.tail(20)
```

### An√°lisis

Una vez proyectada el IPC para los siguientes 6 meses, vamos a utilizar la formula usada por el INE, para el c√°lculo de la inflaci√≥n anual (La f√≥rmula pueden verlo en el siguiente link [F√≥rmula de Calculo INE](https://www.ine.gob.bo/index.php/graficos-ipc/#) )

$$
\text{Inflaci√≥n Anual (%)}_{12\ \text{meses},\ t} = \left( \frac{IPC_t}{IPC_{t - 12}} - 1 \right) \times 100
$$

```python
Base_Total["Inflacion_Anual"]=Base_Total["IPC"].pct_change(periods=12)

Base_Total = Base_Total.dropna(subset=["Inflacion_Anual"])
Base_Total
```

```python
Base_Filtrado = Base_Total[Base_Total.index>"2024-01-01"]
Base_Filtrado
```

```python
fig, ax = plt.subplots(figsize=(15,5))

Fecha_Index = pd.to_datetime("2025-07-01")

Anterior = Base_Filtrado[Base_Filtrado.index < Fecha_Index]
Despues = Base_Filtrado[Base_Filtrado.index >= Fecha_Index]

ax.plot(Anterior.index , Anterior["Inflacion_Anual"],color="black", label="Inflaci√≥n Anual Hist√≥rica")
ax.plot(Despues.index , Despues["Inflacion_Anual"],color="darkred", label="Inflaci√≥n Anual Proyectada")

ax.xaxis.set_major_formatter(mdates.DateFormatter("%b-%y"))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval = 2))

for i,row in Despues.iterrows():
    ax.text(i,row["Inflacion_Anual"]+0.005,f"{row['Inflacion_Anual']:.2%}",ha="center",fontsize=8,color="red")

plt.title("Proyecci√≥n de la inflaci√≥n en Bolivia ‚Äì Pr√≥ximos 6 meses (SARIMA)")
plt.legend()
plt.show()
```
